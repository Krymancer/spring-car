```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def MassSpring(state,t):
  # Descompacta o vetor de stados;
  # unpack the state vector
  x = state[0]
  xd = state[1]

  # Constantes
  k = -2.5 # Newtons por metro
  m = 1.5 # Kilogramas
  g = 9.8 # metros por segundo

  # computa a aceleração xdd
  xdd = ((k*x)/m) + g

  # retorna o estado das duas derivadas
  return [xd, xdd]
#--------------- End of MassSpring ---------------#

state0 = [0.0, 0.0]
t = np.arange(0.0, 10.0, 0.1)

state = odeint(MassSpring, state0, t)

plt.plot(t, state)
plt.xlabel('Tempo (s)')
plt.ylabel('Estados')
plt.title("Sistema Massa-mola")
plt.legend(('$x$ (m)', '$\dot{x}$ (m/s)'))
plt.show()
```
